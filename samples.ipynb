# An example of simple clustering

```python
import networkx as nx
from cw import chinese_whispers, aggregate_clusters

# Create some networkx graph G with two clusters
offset = 100
G = nx.karate_club_graph()
G2 = nx.karate_club_graph()
G.add_nodes_from([n + offset for n in G2.nodes])
G.add_edges_from([(src + offset, dst + offset) for src, dst in G2.edges])

# Perform clustering of G
chinese_whispers(G, weighting="top", iterations=20)

# Print the clusters
print("Cluster ID\tCluster elements\n")
for label, cluster in sorted(aggregate_clusters(G).items(), key=lambda e: len(e[1]), reverse=True):
    print("{}\t{}\n".format(label, cluster))
```

# Clustering and visualizing the result using networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
from chinese_whispers import chinese_whispers

# Create some networkx graph G with two clusters
offset = 100
G = nx.karate_club_graph()
G2 = nx.karate_club_graph()
G.add_nodes_from([n + offset for n in G2.nodes])
G.add_edges_from([(src + offset, dst + offset) for src, dst in G2.edges])

# Perform clustering of G
chinese_whispers(G, weighting="top", iterations=20)

# Visualize the clustering result of G using networkx (requires matplotlib)
colors = [ 1./G.node[node]["label"] for node in G.nodes()]
fig = plt.gcf()
fig.set_size_inches(10, 10)
nx.draw_networkx(G, cmap=plt.get_cmap('jet'), node_color=colors, font_color="white")
plt.show()
```
